## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
import matplotlib.pyplot as plt

# bar chart of total per region


plt.barh(df['Total'],df['Region'] ,color = ['#F0F8FF','#E6E6FA','#B0E0E6']) 

plt.title('Total Score per Region')
plt.xlabel('Region')
plt.ylabel('Total') 

plt.show()
```

```python
# Bar chart to show Status per region - need to redo as this chart is not at all useful

plt.barh(df['Status'],df['Region'],color = ['#F0F8FF','#E6E6FA','#B0E0E6']) 

plt.title('Freedom Status per Region')
plt.xlabel('Status')
plt.ylabel('Region') 

plt.show()
```

```python
Total = df['Total'].sum
data_group = pd.DataFrame(df.loc[['Status'], Total].T)

data_group.plot.bar(edgecolor='white')
plt.title('Total by Region', color= 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.xlabel('Total',color = 'black')
plt.ylabel('Status',color = 'black')
plt.legend(title = '', fontsize = 18, labelcolor='black')
plt.savefig('bar_grouped.png')
style.use('ggplot')

plt.show()
```
